```python
import math
import numpy as np
import matplotlib.pyplot as plt
import random

def plotFuncio(f,xmin,xmax):
    x = np.linspace(xmin,xmax,1000)
    y = [f(i) for i in x]
    plt.plot(x,y); plt.grid(); plt.show()
```

## Cerca àuria
Es defineix la funció $f_1(x)$:

```python
def f1(x):
#     return (x**3)+924*math.exp(-x)
    return -x**3+math.exp(x)
```

I s'executa l'algoritme especificant l'interval d'estudi $[x_1, x_2]$ i el nombre d'iteracions:

```python
def cercaAuria(funcio,x1,x2,iteracions):
    x1_Orig, x2_Orig = x1, x2

    x3 = x2-((math.sqrt(5)-1)/2)*(x2-x1)
    x4 = x1+((math.sqrt(5)-1)/2)*(x2-x1)
    
    print("It. 0    x1 = {}     x2 = {}     x3 = {}     x4 = {}  ".format(round(x1,4),round(x2,4),round(x3,4),round(x4,4)))

    for k in range(1,iteracions+1):
        if funcio(x3) < funcio(x4):
            x1, x2, x3, x4 = x1, x4, x4 - ((math.sqrt(5)-1)/2)*(x4 - x1), x3            
        else:
            x1, x2, x3, x4 = x3, x2, x4, x3 + ((math.sqrt(5)-1)/2)*(x2-x3)
            
        # prints i plots
        print("It. {}    x1 = {}     x2 = {}     x3 = {}     x4 = {}  ".format(k,round(x1,4),round(x2,4),round(x3,4),round(x4,4)))
        plt.plot([x1, x2],[funcio(x1), funcio(x2)],"--",color=(random.random(),random.random(),random.random()))
    plotFuncio(funcio,x1_Orig,x2_Orig)

    
    
# Crida a la funció   
cercaAuria(f1,-1,-0,4)
```

## Bisecció
Es defineix la funció $f_2(x)$ i la seva derivada $f^{'}_2(x)$:

```python
def f2(x):
    return -x**3+math.exp(x)
def Df2(x):
    return -3*x**2+math.exp(x)
```

I s'executa l'algoritme especificant l'interval d'estudi  $[x_1, x_2]$ i el nombre d'iteracions:

```python
def Biseccio(funcio,funcioDerivada,x1,x2,iteracions):
    x1_Orig, x2_Orig = x1, x2
    
    print("It. 0    x1 = {}     x2 = {}".format(x1,x2))

    for k in range(1,iteracions+1):
        if funcioDerivada((x1+x2)/2) > 0:
            if funcioDerivada(x1) < 0:
                x1, x2 = x1, (x1+x2)/2
            elif funcioDerivada(x1) > 0:
                x1, x2 = (x1+x2)/2, x2
        elif funcioDerivada((x1+x2)/2) < 0:
            if funcioDerivada(x1) < 0:
                x1, x2 = (x1+x2)/2, x2
            elif funcioDerivada(x1) > 0:
                x1, x2 = x1, (x1+x2)/2
        else:
            print("òptim!")
            return (x1+x2)/2
            
        # prints i plots
        print("It. {}    x1 = {}     x2 = {}".format(k,round(x1,4),round(x2,4)))
        plt.plot([x1, x2],[funcio(x1), funcio(x2)],"--",color=(random.random(),random.random(),random.random()))
    plotFuncio(funcio,x1_Orig,x2_Orig)

    
    
# avaluem    
Biseccio(f2,Df2,-1,0,4)
```

## Newton
Es defineix la funció $f_3(x)$, la seva derivada primera $f^{'}_3(x)$ i la seva derivada segona $f^{''}_3(x)$:

```python
def f3(x):
    return x**3+799*math.exp(-x)
def Df3(x):
    return 3*x**2-799*math.exp(-x)
def DDf3(x):
    return 6*x+799*math.exp(-x)
```

I s'executa l'algoritme especificant el punt de sortida $x_1$ i el nombre d'iteracions:

```python
def Newton(funcio,funcD,funcDD,x1,iteracions):
   
    print("It. 0    x1 = {}".format(round(x1,4))); plt.scatter(x1,funcio(x1),color=(random.random(),random.random(),random.random()))
    memoryX = [x1]
    
    for k in range(1,iteracions+1):
        x1 = x1 - funcD(x1)/funcDD(x1)
        print("It. {}    x1 = {}".format(k,round(x1,4))); memoryX.append(x1)
                 
        # prints i plots
        plt.scatter(x1,funcio(x1),color=(random.random(),random.random(),random.random()))
    plotFuncio(funcio,min(memoryX),max(memoryX))
    
# avaluem    
Newton(f3,Df3,DDf3,-20,30)
```

## Nelder i Mead
Es defineix la funció $f_4(x)$:

```python
def f4(punt):
    x, y = punt[0], punt[1]
    return -4*x-14*y+x**2+y**2
```

Funcions complementàries per l'algoritme:

```python
def fi(p1,p2,a):
    return p1+a*(p2-p1)

def beBaixa(vg,llista_v):
    # centre de gravetat de la cara de l’hipertetraedre oposada a vg
    v = np.array([0,0])
    for punt in llista_v:
        if np.linalg.norm(punt-vg) > 0.000001:
            v = v + punt
    return (1/(len(llista_v)-1))*v

def puntOrdenatsPetit_a_Gran(funcio,llistPunts):
    llistaImatges = [funcio(punt) for punt in llistPunts]

    indexMax = llistaImatges.index(max(llistaImatges))
    puntMax = llistPunts.pop(indexMax)
    llistaImatges.pop(indexMax)
    
    
    indexMin = llistaImatges.index(min(llistaImatges))
    puntMin = llistPunts.pop(indexMin)
    llistaImatges.pop(indexMin)

    return [puntMin, llistPunts[0], puntMax]

# def puntOrdenatsPetit_a_Gran_v2(funcio,llistPunts):
#     llistaImatges = [funcio(punt) for punt in llistPunts]

#     indexMax = llistaImatges.index(max(llistaImatges))
#     puntMax = llistPunts.pop(indexMax)
#     llistaImatges.pop(indexMax)
    
    
#     indexMin = llistaImatges.index(min(llistaImatges))
#     puntMin = llistPunts.pop(indexMin)
#     llistaImatges.pop(indexMin)
    
#     index2Max = llistaImatges.index(max(llistaImatges))
#     punt2Max = llistPunts.pop(index2Max)

#     return [puntMin, punt2Max, puntMax]

def ordenaTriangle(diccionariTriangle):
    vs, vh, vg = puntOrdenatsPetit_a_Gran(f4,list(diccionariTriangle.values))
    return {"vs": vs, "vh": vh, "vg": vg}
    
def plotIteracio(puntsTriangle):
    col = (random.random(),random.random(),random.random())
    plt.plot([puntsTriangle[0][0],puntsTriangle[1][0]],[puntsTriangle[0][1],puntsTriangle[1][1]],"-",color=col)
    plt.plot([puntsTriangle[1][0],puntsTriangle[2][0]],[puntsTriangle[1][1],puntsTriangle[2][1]],"-",color=col)
    plt.plot([puntsTriangle[2][0],puntsTriangle[0][0]],[puntsTriangle[2][1],puntsTriangle[0][1]],"-",color=col)
    plt.scatter([puntsTriangle[0][0],puntsTriangle[1][0],puntsTriangle[2][0]],[puntsTriangle[0][1],puntsTriangle[1][1],puntsTriangle[2][1]],s=30,color = col)
```

I s'executa l'algoritme especificant tres punts $P_1$, $P_2$ i $P_3$ i el nombre d'iteracions:

```python
def NelderMead(funcio,P1,P2,P3,iteracions):
    # vàlid per R2
    vs, vh, vg = puntOrdenatsPetit_a_Gran(funcio, [P1,P2,P3])
    
    # prints i plots
    print("It. 0    vs = {}   vh = {}   vg = {}".format(vs,vh,vg))
    plotIteracio([vs,vh,vg])
    
    for k in range(1,iteracions+1):
        dicc_triangle = {"vs": vs, "vh": vh, "vg": vg}

        v = beBaixa(vg,[vs,vh,vg])
        print("v = ",v)
        vr = fi(v, vg, -1)
        print("vr = ",f4(vr))
        # CAS 1
        if funcio(vr) < funcio(vs):
            ve = fi(v,vg,-2)
            # CAS 1a
            if funcio(ve) < funcio(vs):
                dicc_triangle["vg"] = ve 
            # CAS 1b
            else:
                dicc_triangle["vg"] = vr
        # CAS 2
        elif funcio(vs) <= funcio(vr) and funcio(vr) < funcio(vh):
            dicc_triangle["vg"] = vr
        # CAS 3
        elif funcio(vh) <= funcio(vr) and funcio(vr) < funcio(vg):
            dicc_triangle["vg"] = vr
            vc = fi(v, dicc_triangle["vg"], 1/2)
            print(vc)
            # CAS 3a
            if funcio(vc) < funcio(dicc_triangle["vg"]):
                dicc_triangle["vg"] = vc
            # CAS 3b
            else:
                dicc_triangle["vh"] = fi(vs, dicc_triangle["vh"], 1/2)
                dicc_triangle["vg"] = fi(vs, dicc_triangle["vg"], 1/2)
        # CAS 4
        else:
            vc = fi(v, dicc_triangle["vg"], 1/2)
            # CAS 3a
            if funcio(vc) < funcio(dicc_triangle["vg"]):
                dicc_triangle["vg"] = vc
            # CAS 3b
            else:
                dicc_triangle["vh"] = fi(vs, dicc_triangle["vh"], 1/2)
                dicc_triangle["vg"] = fi(vs, dicc_triangle["vg"], 1/2)
        
        nouTriangle = list(dicc_triangle.values())
        vs, vh, vg = puntOrdenatsPetit_a_Gran(funcio, nouTriangle)
        fs, fh, fg = funcio(vs), funcio(vh), funcio(vg)
    
    
        plotIteracio([vs,vh,vg])
        print("It. {}    vs = {}   vh = {}   vg = {}".format(k,vs,vh,vg))
        print("It. {}    f(vs) = {}   f(vh) = {}   f(vg) = {}".format(k,funcio(vs),funcio(vh),funcio(vg)))
    plt.grid()


# avaluem

# (exemple de les diapos)
v0 = np.array((8, 5))
v1 = np.array((10, 10))
v2 = np.array((6,12))

NelderMead(f4,v0,v1,v2,23)
```
